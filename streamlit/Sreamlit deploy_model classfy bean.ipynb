```python

%%writefile app.py
import streamlit as st
import pandas as pd
from PIL import Image
import streamlit as st


import streamlit as st
import pandas as pd
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
from tensorflow.keras import preprocessing
from tensorflow.keras.models import load_model
from tensorflow.keras.activations import softmax
import os
import h5py


def main_page():
    st.sidebar.markdown("###### BEAN DISEASE PREDICTION")
    with st.container():
        st.markdown('---')
        st.markdown\
             ("<h1 style='text-align: center; color: #728FCE;'>PLANT DISEASE DETECTION</h1>", unsafe_allow_html=True)
        st.markdown("<h3 style='text-align: center; color: #728FCE;'>(by Silvia Rahma)</h1>", unsafe_allow_html=True)
        st.markdown("<h1 style='text-align: center; color: #728FCE;'>ğŸ¦ âš ï¸ğŸ–¥ï¸</h1>", unsafe_allow_html=True)
        st.markdown('---')
        st.markdown("<h3 style='text-align: center; color: #728FCE;'>Features</h3>", unsafe_allow_html=True)
        st.markdown('- Disease Prediction'
            '\n- Disease Definition'
            '\n- Disease Occurence'
            '\n- Disease Importance'
            '\n- Disease Treatment')
# Kacang
def bean():
    st.sidebar.markdown('- Deteksi penyakit Kacang'
                    '\n- Penjelasan Penyakit Kacang')
    st.markdown('---')
    st.markdown("<h1 style='text-align: center; color: #728FCE;'>Tanaman Kacang</h1>", unsafe_allow_html=True)
    st.markdown("<h1 style='text-align: center; color: #728FCE;'>ğŸ¦ </h1>", unsafe_allow_html=True)
    st.markdown('---')

    # Deteksi penyakit
    st.markdown("<h2 style='color: #728FCE;'>Upload Gambar Daun Kacang</h1>", unsafe_allow_html=True)
    # fungsi untuk import atau upload gambar

    #fungsi untuk prediksi kelas
    def predict_class_kacang(image):
        classifier_model = tf.keras.models.load_model(r'/content/drive/MyDrive/portofolio/streamlit bean/resnet_kacang.hdf5')
        shape = ((224, 224, 3))
        model = tf.keras.Sequential([hub.KerasLayer(classifier_model, input_shape=shape)])

        test_image = image.resize((224, 224))
        test_image = preprocessing.image.img_to_array(test_image)
        #test_image = test_image/225.0
        test_image = np.expand_dims(test_image, axis = 0)
        class_names = ['bean angular leaf spot', 'bean rust', 'bean healthy']
        prediction = model.predict(test_image)
        score = tf.nn.softmax(prediction[0])
        score = score.numpy()
        image_class =  class_names[np.argmax(score)]
        return image_class

    file_uploaded = st.file_uploader("Pilih File", type = ['jpg', 'png', 'jpeg'])
    if file_uploaded is not None:
        image = Image.open(file_uploaded)
        figure = plt.figure(figsize=(4, 3))
        plt.imshow(image)
        plt.axis('off')
        predict_class = predict_class_kacang(image)
        st.markdown("<h2 style='color: #728FCE;'>Prediksi penyakit Kacang : {}</h1>".format(predict_class), unsafe_allow_html=True)
        st.pyplot(figure)

     # Penjelasan Penyakit
        st.markdown("<h2 style='color: #728FCE;'>Penjelasan Penyakit : {}</h1>".format(predict_class), unsafe_allow_html=True)

        if(predict_class=='bean angular leaf spot'):
            st.write('#### Definisi :')
            st.write(" Angular leaf spot merupakan penyakit pada tanaman kacang yang disebabkan oleh jamur Cercospora arachidicola dan Phaeoisariopsis griseola. ")
            st.write('#### Kemunculan :')
            st.write(" Penyakit Angular Leaf spot sering terjadi di daerah dengan suhu 16-28 derajat Celcius, dan pada keadaan tertentu seperti saat cuaca basah atau kelembaban tinggi bergantian dengan kondisi kering dan berangin.")
            st.write('#### Penting:')
            st.write(" penyakit dapat berkembang sangat cepat pada tanaman berumur 40-45 hari")
            st.write('#### Fungisida :')
            st.write(" - KASUMIN 5/75 WP")
            st.write(" - fungisida klorotalonil KELONIL 75 WP")


        elif(predict_class=='bean rust'):
            st.write('#### Definisi :')
            st.write(" Bean Rust atau kacang karat merupakan pada tanaman kacang yang disebabkan oleh Uromyces appendiculatus ")
            st.write('#### Kemunculan :')
            st.write(" Penyakit ini cepat berkang di daerah lembab dengan suhu sedang. ")
            st.write('#### Penting:')
            st.write(" Infeksi yang parah dapat menyebabkan defoliasi. Daun yang sakit kehilangan kapasitas fotosintesis yang mengakibatkan penurunan hasil panen.")
            st.write('#### Penyembuhan:')
            st.write(" - Gunakan bibit umbi kentang yang sehat dan tidak terserang penyakit. ")
            st.write(" - Lakukan pencelupan bibit umbi kentang maksimal selama Â± 3 menit dalam larutan agens hayati Pseudomonas fluorescens (Pf) dengan dosis larutan 1 ml Pf per liter air.")
            st.write(" - Membersihkan sisa tanaman kentang yang terserang penyakit dengan cara melakukan pembakaran  untuk menghindari terjadinya penularan penyakit ke tanaman kentang yang sehat.")
            st.write(" - Lakukan sanitasi pada tanaman kentang dari bagian tanaman yang terserang penyakit, dan tanaman gulma liar yang berpotensi menularkan beberapa penyakit tanaman.")
            st.write(" - Lakukan pelebaran jarak tanam dan penimbunan umbi yang baik untuk mengurangi jumlah spora yang mencapai umbi kentang.")
            st.write(" - Drainase udara untuk  pengeringan dedaunan setiap harinya. ")
            st.write('#### Pupuk :')
            st.write(" - Pupuk KCL ")

        elif(predict_class=='bean healthy'):
            st.write(" Tanaman Kacang sehat")

page_names_to_funcs = {
    "Headline": main_page,
    "Deteksi Penyakit Tanaman Kacang": bean,
}

selected_page = st.sidebar.selectbox("Select a page", page_names_to_funcs.keys())
page_names_to_funcs[selected_page]()
```
